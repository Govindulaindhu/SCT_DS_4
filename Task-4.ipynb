```python
import pandas as pd

# Load dataset
df = pd.read_csv('accident.csv')

# Convert 'Time' to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')

# Extract hour as 'time_of_day'
df['time_of_day'] = df['Time'].dt.hour

# Check
print(df[['Time','time_of_day']].head())
```

```python
print(df.columns)
```

```python
df['time_of_day'].value_counts().plot(kind='bar', title='Accidents by Time of Day')
plt.show()
```

```python
df['Day_of_week'].value_counts().sort_index().plot(kind='bar', title='Accidents by Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['Road_surface_conditions'].value_counts().plot(kind='bar', title='Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['Weather_conditions'].value_counts().plot(kind='bar', title='Accidents by Weather Conditions')
plt.xlabel('Weather')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
sns.histplot(df['Accident_severity'], bins=5, kde=True)
plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
df['Type_of_collision'].value_counts().plot(kind='bar', title='Accidents by Type of Collision')
plt.xlabel('Collision Type')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
sns.boxplot(x='time_of_day', y='Accident_severity', data=df)
plt.title('Accident Severity vs Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Severity')
plt.show()
```

```python
sns.boxplot(x='Day_of_week', y='Accident_severity', data=df)
plt.title('Accident Severity vs Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Severity')
plt.show()
```

```python
sns.boxplot(x='Area_accident_occured', y='Accident_severity', data=df)
plt.title('Area of Accident vs Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Day_of_week', hue='Weather_conditions', data=df)
plt.title('Weather Conditions Across Days')
plt.show()
```

```python
sns.histplot(df['Number_of_casualties'], bins=5, kde=True)
plt.title('Number of Casualties Distribution')
plt.show()
```
